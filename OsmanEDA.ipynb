```python
import pandas as pd

# Load your data
# movies = pd.read_csv('path_to_movies.csv')

# Ensure 'release_date' is in datetime format
movies['release_date'] = pd.to_datetime(movies['release_date'], errors='coerce')

# Filter movies where both budget and revenue are greater than zero and release_date is on or before January 1, 2024
filtered_movies = movies[(movies['budget'] > 0) & (movies['revenue'] > 0) & (movies['release_date'] <= '2023-06-01')]

# Calculate the profit
filtered_movies['profit'] = filtered_movies['revenue'] - filtered_movies['budget']

# Calculate the profit-to-budget ratio
filtered_movies['profit_to_budget_ratio'] = filtered_movies['profit'] / filtered_movies['budget']

# Define categories based on the ratio
def categorize_movie(ratio):
    if ratio < 0:
        return 'Flop'
    elif ratio < 2:
        return 'Success'
    else:
        return 'Blockbuster'

# Apply the function to categorize movies
filtered_movies['success_level'] = filtered_movies['profit_to_budget_ratio'].apply(categorize_movie)

# Print a few rows to confirm the categories are added correctly
print(filtered_movies[['title', 'budget', 'revenue', 'profit', 'profit_to_budget_ratio', 'success_level', 'release_date']].head())

# Count the number of movies in each success level category and print
success_level_counts = filtered_movies['success_level'].value_counts()
print(success_level_counts)

# Calculate the number of movies remaining in the filtered dataset
num_movies = filtered_movies.shape[0]

# Print the number of movies
print(f"The number of movies released on or before June 1, 2023, with non-zero budgets and revenues is: {num_movies}")

```

```python
# Count the number of movies in each success level category
success_level_counts = filtered_movies['success_level'].value_counts()

# Print the counts for each category
print(success_level_counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set(style="whitegrid")


# Bar chart of success levels
plt.figure(figsize=(10, 6))
success_level_counts = filtered_movies['success_level'].value_counts()
sns.barplot(x=success_level_counts.index, y=success_level_counts.values)
plt.title('Count of Movies by Success Level')
plt.xlabel('Success Level')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()

```

```python
# Count the number of movies in each success level category
success_counts = filtered_movies['success_level'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(success_counts, labels=success_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Movie Success Levels')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Ensure the 'release_date' is in datetime format
filtered_movies['release_date'] = pd.to_datetime(filtered_movies['release_date'], errors='coerce')

# Now extract the year and proceed with the analysis
filtered_movies['release_year'] = filtered_movies['release_date'].dt.year

# Aggregate average budget by year
average_budget_by_year = filtered_movies.groupby('release_year')['budget'].mean()

plt.figure(figsize=(12, 6))
plt.plot(average_budget_by_year.index, average_budget_by_year, marker='o')
plt.title('Average Movie Budgets Over Years')
plt.xlabel('Year')
plt.ylabel('Average Budget')
plt.grid(True)
plt.show()
```

```python
# Display movie entries until the year June 1, 2023 to inspect the budget values
recent_movies = filtered_movies[filtered_movies['release_date'] <= '2023-06-01']
print(recent_movies[['title', 'release_year', 'budget']].sort_values(by='release_year'))
```
